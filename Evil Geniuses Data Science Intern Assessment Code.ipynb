```python
import numpy as np
from sklearn.model_selection import KFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Parameter Grid for Grid Search
param_grid = {
    'layer_sizes': [(16,), (32,), (64,), (128,), (64, 32), (128, 64)],
    'dropout_rate': [0.1, 0.2, 0.3],
    'learning_rate': [0.001, 0.01, 0.1]
}


x = xData.values
y = yData
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Loop over the parameter grid and do cross-validation on the models
results = []
for layer_sizes in param_grid['layer_sizes']:
    for dropout_rate in param_grid['dropout_rate']:
        for learning_rate in param_grid['learning_rate']:
            fold_results = []
            for train_idx, val_idx in cv.split(x, y):
                # Create Model
                model = Sequential()
                for layer_size in layer_sizes:
                    model.add(Dense(layer_size, activation='relu'))
                    model.add(Dropout(dropout_rate))
                model.add(Dense(1, activation='linear'))

                # Compile Model, specifically using mean squared error
                optimizer = Adam(learning_rate=learning_rate)
                model.compile(optimizer=optimizer,
                              loss='mean_squared_error',
                              metrics=['mean_squared_error'])

                # Train Model
                early_stop = EarlyStopping(patience=5)
                history = model.fit(x[train_idx], y[train_idx],
                                    validation_data=(x[val_idx], y[val_idx]),
                                    epochs=50, batch_size=64, callbacks=[early_stop])

                # Evaluate Model on Test Set
                test_loss, test_mse = model.evaluate(x, y)
                fold_results.append(test_mse)

            # Compute the mean squared error over the folds and store the results
            mean_mse = np.mean(fold_results)
            results.append((layer_sizes, dropout_rate, learning_rate, mean_mse))

# Print Results
for result in results:
    print('Layer sizes: {}, Dropout rate: {}, Learning rate: {}, Test MSE: {:.4f}'.format(result[0],
                                                                                            result[1],
                                                                                            result[2],
                                                                                            result[3]))
```

```python
# Sort Results by mean squared error to find the best model
results.sort(key=lambda x: x[3])

# Print the best model
print('Best model: Layer sizes: {}, Dropout rate: {}, Learning rate: {}, Test MSE: {:.4f}'.format(
    results[0][0], results[0][1], results[0][2], results[0][3]))
```
